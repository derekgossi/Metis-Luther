```python
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
### CHALLENGE #1

# Import dataframe using pickle
with open('df_w_twitter.p', 'rb') as picklefile:
	movie_data = pickle.load(picklefile)

# Drop NaN values from data and remove outliers
movie_data = movie_data.dropna()[movie_data['NumFollowers'] <= 1000000]

# Add ln terms
movie_data['ln_NumTweetsPerYr'] = np.log(movie_data['NumTweetsPerYr'])
movie_data['ln_Age'] = np.log(movie_data['Age'])
movie_data['ln_AvgGross'] = np.log(movie_data['AvgGross'])
movie_data['ln_NumFollowers'] = np.log(movie_data['NumFollowers'])
movie_data['Ones'] = 1.0
```

```python
from sklearn.linear_model import LinearRegression

# Set factor matrix X and observed values y
X = movie_data[['ln_NumTweetsPerYr', 'ln_Age', 'AvgGross']]
y = movie_data['ln_NumFollowers']

# Run linear model
lm = LinearRegression()
lm.fit(X,y)

# Print summary statistics
print lm.coef_
print lm.intercept_
print lm.score(X,y)

plt.plot(y, lm.predict(X), 'go')
plt.plot(y, y, 'm--', label = 'Ideal Prediction')
plt.show()
```

```python
import statsmodels.formula.api as sm

# Set factor matrix X and observed values y
X = movie_data[['ln_NumTweetsPerYr', 'ln_Age', 'AvgGross', 'Ones']]
X.head()
y = movie_data['ln_NumFollowers']

linmodel = sm.OLS(y,X).fit()

print linmodel.summary()

plt.plot(y, linmodel.predict(X), 'go')
plt.plot(y, y, 'm--', label = 'Ideal Prediction')
plt.show()
```

```python
## GENERATE SOME DATA

# Get a vector of x values for our random dataset
#	Let's say 500 observations randomly in the interval 0 to 1000
x_data = np.random.randint(1,1000,500)

# Parameters for our model
beta_vec = [6]
beta_o = 500
err_std_dev = 2

# Iterate through our x data, log transform it, and determine what our random generated y's are for each.
#	Note that the generator function can take a vector of factors + constant, but here we are only sending it 
#	a single factor: log(x)
y_generated = [ gen_val_from_lin_model( [np.log(x_data[x])], beta_vec, beta_o, err_std_dev ) for x in range(len(x_data)) ]

# Plot random generated data
plt.scatter(x_data, y_generated)
plt.title('Random Data Generated from Model y = B_0 + B_1*log(x) + N(0,5)')
plt.show()
```

```python
## FIT MODELS TO GENERATED DATA

# We start by creating a dataframe with our new generated data
gen_df = pd.DataFrame({'x': x_data, 'y': y_generated})

# Quadratic model

# Add quadratic transformed x to our dataframe
gen_df['x^2'] = gen_df['x']**2

# Inputs to our model
X = gen_df[['x','x^2']]
y = gen_df['y']

# Fit linear model
lm_quad = LinearRegression()
fitted_lm_quad = lm_quad.fit(X,y)
pred_y_quad = fitted_lm_quad.predict(X)

# Summary statistics
print fitted_lm_quad.coef_
print fitted_lm_quad.intercept_
print fitted_lm_quad.score(X,y)

# Plot linear model 
plt.scatter(gen_df['x'],gen_df['y'], color='r', marker='x')
plt.scatter(gen_df['x'],pred_y_quad, color='b', marker='o')
plt.show()
```

```python
# Log model

# Add log transformed x to our dataframe
gen_df['ln_x'] = np.log(gen_df['x'])

# Inputs to our model
X = gen_df[['ln_x']]
y = gen_df['y']

# Fit linear model
lm_log = LinearRegression()
fitted_lm_log = lm_log.fit(X,y)
pred_y_log = fitted_lm_log.predict(X)

# Summary statistics
print fitted_lm_log.coef_
print fitted_lm_log.intercept_
print fitted_lm_log.score(X,y)

# Plot linear model 
plt.scatter(gen_df['x'],gen_df['y'], color='r', marker='x')
plt.scatter(gen_df['x'],pred_y_log, color='b', marker='o')
plt.show()
```
